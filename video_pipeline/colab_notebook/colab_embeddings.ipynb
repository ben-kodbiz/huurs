# 🔢 Generate Embeddings for Processed Chunks

**Purpose:** Generate vector embeddings for semantic search

**Model:** sentence-transformers/all-MiniLM-L6-v2 (fast & lightweight)

**Time:** ~2-3 minutes for 712 chunks

## Step 2: Load Processed Chunks

```python
import json

# Upload processed_chunks.json first!
try:
    with open('processed_chunks.json') as f:
        chunks = json.load(f)
    print(f"✓ Loaded {len(chunks)} processed chunks")
except FileNotFoundError:
    print("⚠️ processed_chunks.json not found!")
    print("Please upload the file from your computer.")
```

## Step 3: Load Embedding Model

```python
from sentence_transformers import SentenceTransformer

# Lightweight model for fast embeddings
MODEL_NAME = "sentence-transformers/all-MiniLM-L6-v2"

print(f"Loading embedding model: {MODEL_NAME}...")
model = SentenceTransformer(MODEL_NAME)
print(f"✓ Model loaded")
```

## Step 4: Generate Embeddings

```python
import numpy as np

# Prepare texts for embedding (combine summary + text)
texts_to_embed = []
for chunk in chunks:
    # Combine summary and topic for better semantic search
    text = f"{chunk.get('summary', '')} {chunk.get('topic', '')} {chunk['text'][:300]}"
    texts_to_embed.append(text)

print(f"Generating embeddings for {len(texts_to_embed)} chunks...")

# Generate embeddings in batches
embeddings = model.encode(texts_to_embed, batch_size=32, show_progress_bar=True)

print(f"✓ Generated {len(embeddings)} embeddings")
print(f"  Embedding dimensions: {len(embeddings[0])}")
```

## Step 5: Save Embeddings with Chunks

```python
# Add embeddings to chunks
for i, chunk in enumerate(chunks):
    chunk['embedding'] = embeddings[i].tolist()

# Save
with open('processed_chunks_with_embeddings.json', 'w', encoding='utf-8') as f:
    json.dump(chunks, f, indent=2, ensure_ascii=False)

print(f"✓ Saved to: processed_chunks_with_embeddings.json")
print(f"  File size: {len(chunks)} chunks with embeddings")
```
